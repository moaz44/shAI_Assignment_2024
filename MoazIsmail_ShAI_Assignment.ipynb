## Task **2 Descriptive Statistics:**
 Calculate basic statistics mean, median, mode, minimum, and maximum salary, determine the range of salaries, and find the standard deviation.

```python
# Replace negative salaries with their absolute values if exists
df['TotalPay'] = df['TotalPay'].abs()

#Delete all rows that contain a salary and TotalPayBenefits equal to zero
df.drop(df[df['TotalPayBenefits'] == 0].index, inplace=True)
df.drop(df[df['TotalPay'] == 0].index, inplace=True)
```

```python
num_rows, num_columns = df.shape

print(f"Number of rows = {num_rows}")
print(f"Number of columns = {num_columns}")
```

```python
# Calculate mean, median, mode, minimum, and maximum salary
mean_salary = df['TotalPay'].mean()
median_salary = df['TotalPay'].median()
mode_salary = df['TotalPay'].mode().iloc[0]
min_salary = df['TotalPay'].min()
max_salary = df['TotalPay'].max()
```

```python
print(f"Mean Salary = {mean_salary:.2f}")
print(f"Median Salary = {median_salary:.2f}")
print(f"Mode Salary = {mode_salary:.2f}")
print(f"Minimum Salary = {min_salary:.2f}")
print(f"Maximum Salary = {max_salary:.2f}")
```

```python
#the range of salaries
salary_range = max_salary - min_salary

#the standard deviation
std_dev_salary = df['TotalPay'].std()

print(f"Range of Salaries = {salary_range:.2f}")
print(f"Standard Deviation of Salaries = {std_dev_salary:.2f}")
```

## Task **3 Data Cleaning:**
Handle missing data by suitable method with explain why you use it.

```python
# Check for missing values before remove it
print("Missing values before removal:")
print(df.isnull().sum())
```

```python
#Remove Unnecessary Columns:
columns_to_delete = ['Notes','Status']
df.drop(columns=columns_to_delete, inplace=True)
print(df.isnull().sum())
```

## Task **4 Basic Data Visualization**:
 Create histograms or bar charts to visualize the distribution of salaries, and use pie charts to represent the proportion of employees in different departments.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#histograms visualize the distribution of salaries

plt.figure(figsize=(12, 6))
sns.histplot(df['TotalPay'], bins=40, kde=True, color='skyblue')
plt.title('Distribution of Salaries')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pie chart for the proportion of employees in different departments
department_counts = df['JobTitle'].value_counts().head(10)  # Displaying the top 10 departments for better readability

plt.figure(figsize=(12, 12))
department_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Proportion of Employees in Different Departments')
plt.ylabel('')  # Remove the default ylabel
plt.show()
```

## Task **5 Grouped Analysis:**
 Group the data by one or more columns and calculate summary statistics for each group, and compare the average salaries across different groups.

```python
grouped_by_year = df.groupby('Year')

# Calculate summary statistics for each group
summary_statistics = grouped_by_year['TotalPay'].describe()


print("Summary Statistics for Average Salaries Across Different Years:")
print(summary_statistics)
```

```python
# calculate the average salary for each group
average_salary_by_year = grouped_by_year['TotalPay'].mean()


print("\nAverage Salaries Across Different Years:")
print(average_salary_by_year)
```

## Task **6 Simple Correlation Analysis:**
 Identify any correlation between salary and another numerical column, and plot a scatter plot to visualize the relationship.

```python
# Identify correlation between 'TotalPay' and another numerical column
correlation = df['TotalPay'].corr(df['OtherPay'])

# Print the correlation coefficient
print(f"Correlation between TotalPay and OtherPay: {correlation}")

# Plot a scatter plot to visualize the relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(x='OtherPay', y='TotalPay', data=df, color='coral', alpha=0.7)
plt.title('Scatter Plot: TotalPay vs OtherPay')
plt.xlabel('OtherPay')
plt.ylabel('TotalPay')
plt.show()
```
